### A Simple Neural Network Using Tensorflow

```python
import numpy as np
import tensorflow as tf
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
input_node = 3
hidden_layer_node = 2
output_node = 1
```

```python
w = { 'h1' : tf.Variable(tf.ones([input_node, hidden_layer_node])),
     'w_output' : tf.Variable(tf.ones([hidden_layer_node, output_node]))}

b = { 'b1' : tf.Variable(tf.zeros([hidden_layer_node])),
        'b_output' : tf.Variable(tf.zeros([output_node]))}
```

```python
x = tf.placeholder('float', [None, input_node])
y = tf.placeholder('float', [None, output_node])
```

```python
l1 = tf.add(tf.matmul(x, w['h1']), b['b1'])
l1_activation = tf.sigmoid(l1)
```

```python
output = tf.add(tf.matmul(l1_activation, w['w_output']), b['b_output'])
output_activation = tf.sigmoid(output)
```

```python
cost = tf.reduce_mean(tf.abs(tf.subtract(output_activation, y)))
```

```python
#optimizer = tf.train.AdadeltaOptimizer(learning_rate=0.01).minimize(cost)
optimizer = tf.train.AdamOptimizer(learning_rate=0.01).minimize(cost)
```

```python
x_raw = np.array([[1, 2,3]])
y_raw = np.array([[3]])
```

```python
epochs_number=10000
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    prediction = output_activation.eval({x:x_raw})
    print(prediction)
    
    #For Test Purpose Only
    print(sigmoid(2*sigmoid(6)))

    
    for epoch in range(epochs_number):
        _,epoch_loss = sess.run([optimizer, cost], feed_dict={x: x_raw, y: y_raw})
        print('Epoch', epoch, 'completed out of', epochs_number, 'loss:', epoch_loss)
        
```
